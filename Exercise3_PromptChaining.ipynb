```python
def self_reflection_agent(prompt: str) -> str:
    """
    Simulates summary -> self-critique -> revised summary.
    """
    print(f"--- Sending Prompt to AI Agent ---\n{prompt}\n----------------------------------")

    p = prompt.lower()

    # CHECK CRITIQUE FIRST
    if "self-critique" in p or "critique" in p:
        return (
            "Critique:\n"
            "- Missing the exact 35% wait-time reduction.\n"
            "- Missing the 6-week timeline.\n"
            "- Missing privacy compliance mention.\n"
            "- Too vague and lacks measurable results.\n"
            "- Does not clearly meet the 60–80 word requirement.\n\n"
            "Revised summary:\n"
            "After deploying an AI triage system, customer support wait times dropped by 35% and satisfaction increased from 4.1 to 4.5 out of 5. "
            "The rollout took 6 weeks and included staff training, knowledge base updates, and weekly quality monitoring. "
            "The biggest challenge was handling edge cases while maintaining privacy compliance, but the result was faster and more consistent support for customers."
        )

    # BEFORE summary
    if "before summary" in p:
        return (
            "The company deployed an AI system to improve customer support. "
            "The rollout included training and knowledge updates, and it improved customer satisfaction. "
            "There were challenges, but overall it helped the support process."
        )

    return "Simulated response."
```

```python
source_text = (
    "Our app reduced customer support wait times by 35% after deploying an AI triage system. "
    "The rollout took 6 weeks and involved training staff, updating the knowledge base, and "
    "monitoring quality weekly. Customer satisfaction increased from 4.1 to 4.5 out of 5. "
    "The biggest challenge was handling edge cases and ensuring privacy compliance."
)

requirements = """
Requirements:
- Audience: non-technical manager
- Length: 60–80 words
- Must include: 35% improvement, 6-week timeline, privacy compliance
- Tone: clear and confident
- Format: one paragraph, no bullets
"""

# BEFORE summary
before_prompt = f"""
BEFORE SUMMARY:
Write a summary of the text below.

{requirements}

Text:
{source_text}
"""
before = self_reflection_agent(before_prompt)
print("BEFORE SUMMARY:\n", before)

# Self-critique + revise
critique_prompt = f"""
SELF-CRITIQUE / CRITIQUE-AND-REVISE:
Critique the BEFORE summary against the requirements, then write an improved summary that meets ALL requirements.

Criteria:
- Accuracy (no invented facts)
- Coverage (must include 35%, 6 weeks, privacy compliance)
- Audience fit (non-technical manager)
- Length (60–80 words)
- Clarity (no jargon)
- Format (one paragraph, no bullets)

Text:
{source_text}

BEFORE summary:
{before}

Output format:
1) Critique: 3–6 bullet points
2) Revised summary: one paragraph
"""
after = self_reflection_agent(critique_prompt)
print("\nSELF-CRITIQUE + REVISED SUMMARY:\n", after)
```
